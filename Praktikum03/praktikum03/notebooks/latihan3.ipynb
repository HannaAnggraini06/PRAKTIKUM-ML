```python
import pandas as pd
from sklearn.model_selection import train_test_split

#baca dataset
df = pd.read_csv(path + '/data/day.csv')

#menampilkan 5 data pertama
df.head()
```

```python
df.describe()
```

```python
X = df[['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 'weathersit', 'temp', 'atemp', 'hum', 'windspeed']]

Y = df['cnt']

df.head()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ambil hanya kolom numerik
df_numeric = df.select_dtypes(include=['number'])

# Hitung korelasi
corr_matrix = df_numeric.corr()

# Tampilkan hasil
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix[['cnt']].sort_values(by='cnt', ascending=False),
            annot=True, cmap='Blues')

plt.title('Korelasi Variabel terhadap cnt')
plt.show()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

print("Jumlah data train : ", len(X_train))
print("Jumlah data test : ", len(X_test))

X_train.head()
```

```python
import statsmodels.api as sm

X_train_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train_const).fit()
print(model.summary())
```

```python
X_test_const = sm.add_constant(X_test)

y_pred_test = model.predict(X_test_const)

hasil = pd.DataFrame({'cnt_asli': y_test, 'cnt_pred': y_pred_test})

hasil.head()
```
