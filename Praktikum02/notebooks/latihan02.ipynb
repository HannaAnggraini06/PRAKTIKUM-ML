```python
import pandas as pd
from sklearn.model_selection import train_test_split

#baca dataset
df = pd.read_csv(path + '/data/day.csv')

#menampilkan 5 data pertama
df.head()
```

```python
#split : Data Training 80%, Data Ttesting 20%
train, test = train_test_split(df, test_size=0.2, random_state=42)

#dari Data Training, 10% Data Validation
train, val = train_test_split(train, test_size=0.1, random_state=42)

print("Total data : " , len(df))
print("Jumlah Data Training: " , len(train))
print("Jumlah Data Validation: " , len(val))
print("Jumlah Data Testing: ", len(test))

print("\nContoh Data Training: ")
display(train.head())

print("\nContoh Data Validation: ")
display(val.head())

print("\nContoh Data Testing: ")
display(test.head())
```
